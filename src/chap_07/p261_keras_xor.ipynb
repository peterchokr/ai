### XOR를 학습하는 MLP

필요한 라이브러리 포함

```python
import numpy as np
import tensorflow as tf
```

순방향 신경망 모델. Sequential 생성

```python
model = tf.keras.models.Sequential()
```

모델에 레이어 추가. Dense 클래스는 완전연결 레이어(fully connected layer)를 생성한다.

```python
model.add(tf.keras.layers.Dense(units=2, input_shape=(2,), activation='sigmoid')) #①
model.add(tf.keras.layers.Dense(units=1,  activation='sigmoid')) #②	
```

학습과정 컴파일

```python
model.compile(loss='mean_squared_error',
              optimizer=tf.keras.optimizers.SGD(lr=0.3)) 
```

```python
model.summary()
```

XOR에 대하여 학습 수행

```python
X = np.array([[0, 0],[0, 1],[1, 0],[1, 1]])     # 학습데이터는 반드시 2차원 행렬로.
y = np.array([[0], [1], [1], [0]])
```

```python
model.fit(X, y, batch_size=1, epochs=10000)   #batch_size=1 : 한 샘플마다 가중치 변경
```

```python
print( model.predict(X) )   # epochs=10000 에서는 예측결과가 만족할 만한 수준이 나오지 못한다.-> 횟수 증가(25000)
```
