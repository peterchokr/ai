#### p.93 넘파이로 평규 제곱 오차 계산하기

```python
import numpy as np
y = np.array([1, 1, 2, 2, 4])
ypred = np.array([0.7, 1.34, 1.89, 2.53, 3.2])

import matplotlib.pyplot as plt
x = [i for i in range(1, 6)]
plt.plot(x, y, label="y")
plt.plot(x, ypred, label="ypred")
plt.xlabel('X')
plt.ylabel('Value')
plt.legend(loc="upper left")
plt.show()

mse = (np.square(ypred - y)).mean()
print(mse)
```

#### 맵플롯립으로 시그모이드 함수를 그려보자

```python
import matplotlib.pyplot as plt 
import numpy as np 
import math 
  
X = np.linspace(-10, 10, 100) 
Y = 1/(1 + np.exp(-X)) 
  
plt.plot(X, Y) 
plt.xlabel("x") 
plt.ylabel("Sigmoid(X)") 
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    s=1/(1+np.exp(-x))
    ds=s*(1-s)  		# 시그모이드 함수 1차 미분 함수
    return s,ds

X = np.linspace(-10, 10, 100) 
Y1, Y2 = sigmoid(X)
  
plt.plot(X, Y1, X, Y2) 
plt.xlabel("x") 
plt.ylabel("Sigmoid(X), Sigmoid'(X)") 
plt.show() 
```
