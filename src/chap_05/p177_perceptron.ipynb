### 퍼셉트론 학습 알고리즘 구현하기

방법 2 - sklearn에 구현되어 있는 퍼셉트론 클래스를 이용

```python
from sklearn.linear_model import Perceptron
```

```python
epsilon = 0.0000001		# 부동소수점 오차 방지
```

```python
# 논리적 AND 연산 샘플과 정답
X = [[0,0], [0,1], [1,0], [1,1]]
y = [0, 0, 0, 1]
```

```python
# 퍼셉트론을 생성. tol은 종료조건, random_state는 난수의 시드
clf = Perceptron(tol=1e-3, random_state=0)
```

```python
# 학습을 수행
clf.fit(X, y)
```

```python
# 테스트를 수행
print(clf.predict(X))
```

```python
# 분석 결과 평가 실시
import sklearn.metrics as mt
score = mt.accuracy_score(y, clf.predict(X))
print(score)
```
